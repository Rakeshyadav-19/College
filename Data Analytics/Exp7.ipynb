```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("/home/rax/Documents/GIT/Sems/All Sems/Data Analytics/CSV/Exp7.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel('area(sq.ft)')
plt.ylabel('price(sq.ft)')
plt.scatter(df.Area,df.Price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
%matplotlib inline
plt.xlabel('area(sq.ft)')
plt.ylabel('price(sq.ft)')
plt.scatter(df.Area,df.Price,color='red',marker='+')
plt.plot(df.Area,reg.predict(df[['Area']]),color='blue')
```

Cacluate the value of reg.coeff and reg.intercept

```python
reg.coef_
reg.intercept_
reg.predict(np.array(3300).reshape(-1, 1))
```
